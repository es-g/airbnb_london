```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas
import Data_cleaning as t

raw_listings = pd.read_csv('data/listings.csv')
```

```python
listings = t.clean_dataset(raw_listings)
```

# Exploratory Data Analysis

# What boroughs are rated best during pandemic?
For this metric, we will drop the rows that have not received the reviews. We will also visualise the ratings using geographical data with geopandas library

```python
def clean_boroughs(df):
    scores_boroughs = pd.concat([listings.review_scores_rating, listings['neighbourhood_cleansed']], axis=1)
    scores_boroughs.rename(columns={'neighbourhood_cleansed': 'borough'}, inplace=True)
    scores_boroughs = scores_boroughs.dropna(axis=0)
    return scores_boroughs

scores_boroughs = clean_boroughs(listings)
# Importing the London borough boundary GeoJSON file as a dataframe in geopandas
geo_df = geopandas.read_file('data/neighbourhoods.geojson')
geo_df.head()

scores_boroughs.head()
```

```python
len(scores_boroughs[scores_boroughs['review_scores_rating'] > 90]) / scores_boroughs.shape[0] # More than 74% of the reviews are rated above 90%
```

## Distribution of reviews
Before diving into the ratings of boroughs, let's first visualise the distribution of the reviews

```python
scores_boroughs['review_scores_rating'].plot(kind='hist', figsize=(12, 8), density=False, bins=9)
```

It can be seen that most of the reviews are higher than 90 (more than 74% of the reviews are rated higher than 90%). The distribution is left-skewed. 

```python
# Creating a dataframe of listings by mean 
boroughs_mean_rating = pd.DataFrame(scores_boroughs.groupby('borough').mean())
boroughs_mean_rating.rename(columns={'review_scores_rating': 'mean_rating'}, inplace=True)
# Joining the dataframes
boroughs_mean_rating = geo_df.set_index('neighbourhood').join(boroughs_mean_rating)
# Use ggplot style because it looks cool ;)
plt.style.use('ggplot')


# Heatmap
fig, ax = plt.subplots(1, figsize=(10, 8))
boroughs_mean_rating.plot(column='mean_rating', cmap='Blues_r', ax=ax)
ax.axis('off')
# ax.set_title('Mean rating of Airbnb listings in each London borough', fontsize=12)
sm = plt.cm.ScalarMappable(cmap='Blues_r', 
                           norm=plt.Normalize(
                               vmin=min(boroughs_mean_rating.mean_rating), 
                               vmax=max(boroughs_mean_rating.mean_rating))
                          )
sm._A = [] # Creates an empty array for the data range
cbar = fig.colorbar(sm, ax=ax, orientation='vertical')
plt.show()

# Bar plot for borough ratings
ax2 = boroughs_mean_rating['mean_rating'].plot(kind='barh', figsize=(12, 15), color='#86bf91', zorder=2, width=0.85)
plt.xlim(min(boroughs_mean_rating.mean_rating) - 1, max(boroughs_mean_rating.mean_rating) + 1)
plt.show()
```

People love their AirBnb. We can see that all boroughs, on average, are rated higher than 90%. The difference in ratings between boroughs is insignificant. Kingston upon Thames receives the highest location scores. City of London was rated the lowest, on average.

# What are the most expensive areas in London during pandemic?

```python
#Group by borough and find median price
median_prices = listings[['price', 'neighbourhood_cleansed']].groupby(by='neighbourhood_cleansed').median()

# Joining the dataframes
boroughs_median_price = geo_df.set_index('neighbourhood').join(median_prices)
boroughs_median_price = boroughs_median_price.rename(columns={'price': 'median_price'})

# Plot heatmap of prices
fig3, ax3 = plt.subplots(1, figsize=(15, 10))
boroughs_median_price.plot(column='median_price', cmap='Purples_r', edgecolor='0.8', ax=ax3)
ax3.axis('off')

ax3.set_title('Median price of Airbnb listings in each London borough', fontsize=14)
sm = plt.cm.ScalarMappable(cmap='Purples_r', 
                           norm=plt.Normalize(
                               vmin=min(boroughs_median_price.median_price), 
                               vmax=max(boroughs_median_price.median_price))
                          )
sm._A = []
cbar = fig3.colorbar(sm)
plt.show()


# Bar plot
ax2 = boroughs_median_price['median_price'].plot(kind='barh', figsize=(12, 15), color='purple', zorder=2, width=0.85)
plt.xlim(min(boroughs_median_price.median_price) - 5, max(boroughs_median_price.median_price) + 5)
plt.show()
```

Central areas are typically more expensive than outer London. However, it is interesting to note that are some expensive listings to the west of the city along Thames. It should come as no surprise that Kensington and Chelsea were found to be the most expensive areas in London. These neighbourhoods are home to the richest - celebrities, entrepreneurs, all live or have lived here. The average property price of a home in this neghborhood is about $1.4 million. 

# How popular has AirBnB become in London? And how COVID-19 impacted the demand?

```python
def plot_demand(df):
    df['first_review'] = pd.to_datetime(df['first_review'])
    plt.style.use('fivethirtyeight')
    first_review_resampled = df.set_index('first_review').resample('M').size()

    plt.figure(figsize=(16, 9))
    # plt.scatter(first_review_resampled.index, first_review_resampled)
    first_review_resampled.plot(label='Listings getting their first review', color='black')
    plt.title('Listings getting their first review in each month')

    plt.xlabel('')
    plt.ylabel('Number of new listings')
    plt.axvline(pd.to_datetime('2020-03-11'), color='red', linewidth=2)
    plt.show()
    
plot_demand(raw_listings)
```

•	The number of listings increased over the years. We can see almost exponential growth in customer reviews on new listings which indicates a growth in demand

•	We can also observe the seasonal nature of the demand. There are peaks and drops every year indicating that certain months are busier than the others

•	We see a sharp decline in demand in the end of first quarter of 2020. The vertical line shows the date when WHO announced COVID-19 outbreak as a pandemic

## How the demand changed in 2020?
Let's zoom in on 2019 and 2020. 

```python
from matplotlib.dates import DateFormatter

def plot_demand_last_yrs(df):
    plt.style.use('fivethirtyeight')
    first_review_2019 = df[df['first_review'].dt.year == 2019].set_index('first_review').resample('W').size()
    first_review_2020 = df[df['first_review'].dt.year == 2020].set_index('first_review').resample('W').size()

    #Remove 10 rows to be able to plot two lines superimposed
    first_review_2019 = first_review_2019.iloc[:-10]
    months = first_review_2020.index
    fig, ax = plt.subplots(figsize=(15, 10))
    ax.plot(months, first_review_2019, label='2019', color='green')
    ax.plot(months, first_review_2020, label='2020', color='blue')
    ax.axvline(pd.to_datetime('2020-03-11'), color='red', linewidth=2, label='WHO announces pandemic')
    # set y-axis label
    ax.set_ylabel("Number of new listings", color="black", fontsize=12)
    ax.tick_params(axis='y', colors="black", labelsize=12)
    
    ax.set(xlabel="",
           ylabel="Number of new listings",
           title="Listings getting their first review in each week: 2019 vs 2020")

    date_form = DateFormatter("%b")
    ax.xaxis.set_major_formatter(date_form)
    ax.legend(loc='upper right')
    ax2.legend()
    plt.show()

plot_demand_last_yrs(raw_listings)
```

```python
from matplotlib.dates import DateFormatter

def plot_demand_2020(df):
    plt.style.use('fivethirtyeight')
    
    # Daily confirmed cases from gov.uk
    daily_cases = pd.read_csv('data/data_2020-daily_cases.csv') #Read data
    daily_cases['date'] = pd.to_datetime(daily_cases['date']) #Convert to datetime
    daily_cases = daily_cases[['date', 'newCasesBySpecimenDate']].sort_values(by=['date']).reset_index() #Sort and reset index
    daily_cases = daily_cases[daily_cases['date'] <= pd.to_datetime('2020-10-26')] #
    daily_cases_resampled = daily_cases.set_index('date').newCasesBySpecimenDate.resample('w').sum()

    first_review_2020 = df[df['first_review'].dt.year == 2020].set_index('first_review').resample('W').size()
    months = first_review_2020.index

    fig, ax = plt.subplots(figsize=(15, 10))
    ax.plot(months, first_review_2020, label='number of new listings', color='#006400', linestyle='dotted')

    ax.axvline(pd.to_datetime('2020-03-11'), color='red', linewidth=2, label='WHO announces pandemic')
    ax.axvline(pd.to_datetime('2020-03-26'), color='g', linewidth=2, label='the Lockdown Regulations (SI 350)')
    ax.axvline(pd.to_datetime('2020-06-13'), color='b', linewidth=2, label='the Lockdown Regulations (SI 2020/588)')
    ax.axvline(pd.to_datetime('2020-06-19'), color='m', linewidth=2, label='COVID-19 Alert Level is lowered from Level 4 to Level 3')
    ax.axvline(pd.to_datetime('2020-09-14'), color='g', linewidth=2, label='(Coronavirus, Restrictions) (No. 2) (England) (Amendment) (No. 4) Regulations 2020')
    ax.axvline(pd.to_datetime('2020-09-24'), color='b', linewidth=2, label='New regulations (SI 1029)')
    ax.axvline(pd.to_datetime('2020-10-14'), color='grey', linewidth=2, label='First COVID-19 tier regulations')
      
    
    # set y-axis label
    ax.set_ylabel("Number of new listings", color="#006400", fontsize=12)
    ax.tick_params(axis='y', colors="#006400", labelsize=12)
    
    ax2=ax.twinx()
    # make a plot with different y-axis using second axis object
    ax2.plot(months, daily_cases_resampled, color="#778899", linestyle='dashed', label='New confirmed cases weekly')
    ax2.set_ylabel("New confirmed cases weekly", color="#778899", fontsize=12)
    # Set title and labels for axes
    ax.set(xlabel="",
           ylabel="Number of new listings",
           title="Listings getting their first review in each week in 2020")

    date_form = DateFormatter("%b")
    ax.xaxis.set_major_formatter(date_form)
    ax.legend(loc='upper center')
#     ax2.legend()
    plt.show()

plot_demand_2020(raw_listings)
```

There is a big difference in number of new listings between 2019 and 2020 from mid-March to September. The growth has not yet recovered to the level of the previous year.
Following pandemic announcement by WHO and local lockdown regulations, the demand has dropped significantly. The new listings started to appear on Airbnb a few weeks after new lockdown regulations come into effect and UK’s COVID-19 Alert level is lowered from Level 4 (severe risk, high transmission) to Level 3 (substantial risk, general circulation).


# How the number of active listings changed following pandemic announcement?

```python
def plot_active_listings(df):
    from matplotlib.dates import DateFormatter
    # Daily confirmed cases from gov.uk
    daily_cases = pd.read_csv('data/data_2020-daily_cases.csv') #Read data
    daily_cases['date'] = pd.to_datetime(daily_cases['date']) #Convert to datetime
    daily_cases = daily_cases[['date', 'newCasesBySpecimenDate']].sort_values(by=['date']).reset_index() #Sort and reset index
    daily_cases = daily_cases[daily_cases['date'] <= pd.to_datetime('2020-10-24')] #
    
    df['last_review'] = pd.to_datetime(df['last_review'])
    df = df[df['last_review'] >= pd.to_datetime('2020-01-11')]
    
    last_review_2020 = df[df['last_review'].dt.year == 2020].set_index('last_review').resample('w').size()
    daily_cases_resampled = daily_cases.set_index('date').newCasesBySpecimenDate.resample('w').sum()
    months = last_review_2020.index
    print(daily_cases['date'].min(), df[df['last_review'].dt.year == 2020]['last_review'].min())
    print(df[df['last_review'].dt.year == 2020]['last_review'].max())
    
    plt.style.use('fivethirtyeight')
    fig, ax = plt.subplots(figsize=(15, 10))
    ax.plot(months, last_review_2020, color='#006400', label='Listings getting their last review', linestyle='dotted')
    
    ax.axvline(pd.to_datetime('2020-03-11'), color='red', linewidth=2, label='WHO announces pandemic')
    ax.axvline(pd.to_datetime('2020-03-26'), color='g', linewidth=2, label='the Lockdown Regulations (SI 350)')
    ax.axvline(pd.to_datetime('2020-06-13'), color='k', linewidth=2, label='the Lockdown Regulations (SI 2020/588)')
    ax.axvline(pd.to_datetime('2020-06-19'), color='m', linewidth=2, label='COVID-19 Alert Level is lowered from Level 4 to Level 3')
    ax.axvline(pd.to_datetime('2020-09-14'), color='violet', linewidth=2, label='(Coronavirus, Restrictions) (No. 2) (England) (Amendment) (No. 4) Regulations 2020')
    ax.axvline(pd.to_datetime('2020-09-24'), color='b', linewidth=2, label='New regulations (SI 1029)')
    ax.axvline(pd.to_datetime('2020-10-14'), color='grey', linewidth=2, label='First COVID-19 tier regulations')
    ax.legend()
    # set y-axis label
    ax.set_ylabel("Number of active listings", color="#006400", fontsize=12)
    ax.tick_params(axis='y', colors="#006400", labelsize=12)
    
    ax2=ax.twinx()
    # make a plot with different y-axis using second axis object
    ax2.plot(months, daily_cases_resampled, color="#778899", linestyle='dashed')
    ax2.set_ylabel("New confirmed cases weekly", color="#778899", fontsize=12)

    ax2.tick_params(axis='y', colors="#778899", labelsize=12)

    plt.axvline(pd.to_datetime('2020-03-11'), color='#8B0000', linewidth=1, label='WHO announces pandemic')
    date_form = DateFormatter("%b")
#     plt.legend(loc='upper center')
    plt.show()

plot_active_listings(raw_listings)

```

The figure shows that once pandemic has been announced and local lockdown regulations have been implemented, the number of active listings has dropped significantly almost reaching zero at the end of June. 
Following new lockdown regulations where the rules on gatherings are relaxed and lowering UK's COVID-19 Alert Level from Level 4 to Level 3, we observe the number of active listings increase a few weeks later.
The number of active listings started recovering at the end of July and peaked in October followed by seasonal decline. The number of properties getting their last reviews peaked in beginning of October followed by sharp drop. 
Interestingly, alarming increase in number of new confirmed cases did not stop people from renting the listings. In fact, we see that when new confirmed cases started growing exponentially, number of active listings increased as well, reaching its peak in beginning of October.
